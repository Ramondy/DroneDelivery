```python
import pulp
```

```python
# Initialize Class
model = pulp.LpProblem("Maximize Glass Co. Profits", pulp.LpMaximize)

# Define Decision Variables
wine = pulp.LpVariable('Wine', lowBound=0, upBound=6, cat="Integer")
beer = pulp.LpVariable('Beer', lowBound=0, upBound=None, cat="Integer")

# Define Objective Function
model += 5 * wine + 4.5 * beer

# Define constraints
model += 6 * wine + 5 * beer <= 60
model += 10 * wine + 20 * beer <= 150

# Solve Model
model.solve()
print("Produce {} batches of wine glasses".format(wine.varValue))
print("Produce {} batches of beer glasses".format(beer.varValue))
```
